**Regressão Linear**

É um método estátistico de machine learning que podemos usar para quantificar e fazer prediçoes com base em relacionamentos numéricos entre variáveis.

Existem 2 tipos de regressão:

* Regressão Linear simples - um preditor e um predito.

* Regressão Linear multipla - vários preditores e um predito.

Alguns casos que se utilizam a regressão linear

1 - Predição de vendas

2 - Predição de consume de recursos

3 - predição de custo de suprimentos

4 - Predição de ciclo de vida de serviço de telecomunicação

Pressupostos

Para aplicar a regressão todas as variáveis são contínuas, numéricas, não categóricas.

O conjunto de dados não deve apresentar outliers ou valores em branco.

Precisa existir uma relação linear entre o preditor e o predito.

Todos os preditores são independentes.

Residual (ou erros de predição) tem distribuição normal


```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from matplotlib import rcParams
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```
%matplotlib inline
rcParams['figure.figsize'] = 5,4
sns.set_style('whitegrid')

caminho = pd.read_csv('https://raw.githubusercontent.com/BigDataGal/Python-for-Data-Science/master/enrollment_forecast.csv')
caminho.head()
```

**Regressão Linear nos dados de inscrição**

```
sns.pairplot(caminho);
```

```
print(caminho.corr())
```

```
inscritos_col = ['unem','hgrad']
inscritos_data = caminho[inscritos_col].values

inscritos_target = caminho['roll'].values
X,y = scale(inscritos_data), inscritos_target
```

**Conferindo se temos valores nulos**

```
em_branco = X == np.NaN
X[em_branco == True]
```

O Array vazio significa que não tem nenhum valor em branco

```
lin_reg = LinearRegression(normalize= True)

lin_reg.fit(X,y)
```

```
print(lin_reg.score(X,y))
```
